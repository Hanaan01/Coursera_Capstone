```python
explore = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
explored_df = vdf.loc[:,explore]
```

```python
explored_df.head()
```

```python
def clean_category(cat):
    name = cat[0]['name']
    return name
```

```python
get_name = lambda x: clean_category(x)
```

```python
category = explored_df['venue.categories'].apply(get_name)
```

```python
explored_df['venue.categories'] = category
```
